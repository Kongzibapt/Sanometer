```python
from numpy import loadtxt
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Concatenate
from tensorflow.keras import Input
from tensorflow.keras.models import Model
from matplotlib import pyplot as plt
```

```python
# load the dataset
df = pd.read_excel (r'patient_data.xlsx', sheet_name='Sheet1')
raw_datas = df.values.tolist()
print(raw_datas[0],raw_datas[1])
print(df.columns.tolist())

raw = [list(i) for i in zip(*raw_datas)]
print(raw)

features_list = list(df.to_dict().keys())[:-5] 
features_res_list = list(df.to_dict().keys())[-5:] 
print("features : \n", features_list)
print("features res: \n", features_res_list)


input_x = []
for i in range(len(features_list)) :
    input_x.append((df[features_list[i]].tolist()))
   
    
print(len(input_x[0]))
```

```python
train,test = train_test_split(raw_datas, test_size=0.2, random_state = 1)
train,val = train_test_split(train, test_size=0.2, random_state = 1)

def preprocess(train) :
    # split into input (X) and output (y) variables
    X=[]
    y=[]
    y_element1 = []
    y_element2 = []
    y_element3 = []
    y_element4 = []
    y_element5 = []

    for j in range(len(train)):
        X_element = []

        for i in range(len(train[0])):

            if (i<=len(train[0])-6):
                ##on est dans les features
                X_element.append(train[j][i])

            elif (i==len(train[0])-1) :
                if (train[j][i]=='Y') :

                    y_element5.append(1)
                else :
                    ##par défaut si on a pas de réponse je met 0 comme pour 'N'
                    y_element5.append(0)
            elif (i==len(train[0])-2) :
                if (train[j][i]=='Y') :

                    y_element4.append(1)
                else :
                    ##par défaut si on a pas de réponse je met 0 comme pour 'N'
                    y_element4.append(0)
            elif (i==len(train[0])-3) :
                if (train[j][i]=='Y') :

                    y_element3.append(1)
                else :
                    ##par défaut si on a pas de réponse je met 0 comme pour 'N'
                    y_element3.append(0)
            elif (i==len(train[0])-4) :
                if (train[j][i]=='Y') :

                    y_element2.append(1)
                else :
                    ##par défaut si on a pas de réponse je met 0 comme pour 'N'
                    y_element2.append(0)
            elif (i==len(train[0])-5) :
                if (train[j][i]=='Y') :

                    y_element1.append(1)
                else :
                    ##par défaut si on a pas de réponse je met 0 comme pour 'N'
                    y_element1.append(0)


        X.append(X_element)   
    X=np.array(X)
    y.append(np.array(y_element1))  
    y.append(np.array(y_element2))
    y.append(np.array(y_element3))
    y.append(np.array(y_element4))
    y.append(np.array(y_element5))

    return X,y

train_X,y_train = preprocess(train)
test_X,y_test = preprocess(test)
val_X,y_val = preprocess(val)

train_Y = y_train[0] #on teste que avec le premier cas pour l'instant
test_Y = y_test[0]
val_Y = y_val[0]
```

```python
# define the keras model
#dimension = len(X[0])

inputs = Input(shape=(9,))
inputs_2 = Input(shape=(9,))
inputs_3 = Input(shape=(9,))
inputs_4 = Input(shape=(9,))
inputs_5 = Input(shape=(9,))
inputs_6 = Input(shape=(9,))
inputs_7 = Input(shape=(9,))
inputs_8 = Input(shape=(9,))
inputs_9 = Input(shape=(9,))

# the first branch operates on the first input
x = Dense(8, activation="relu")(inputs)
#x = Dense(4, activation="relu")(x)
#x = Model(inputs=inputs, outputs=x)

'''
# the second branch opreates on the second input
y = Dense(8, activation="relu")(inputs_2)
y = Dense(4, activation="relu")(y)
y = Model(inputs=inputs_2, outputs=y)


a = Dense(8, activation="relu")(inputs_3)
a = Dense(4, activation="relu")(a)
a = Model(inputs=inputs_3, outputs=a)


b = Dense(8, activation="relu")(inputs_4)
b = Dense(4, activation="relu")(b)
b = Model(inputs=inputs_4, outputs=b)


c = Dense(8, activation="relu")(inputs_5)
c = Dense(4, activation="relu")(c)
c = Model(inputs=inputs_5, outputs=c)


d = Dense(8, activation="relu")(inputs_6)
d = Dense(4, activation="relu")(d)
d = Model(inputs=inputs_6, outputs=d)


e = Dense(8, activation="relu")(inputs_7)
e = Dense(4, activation="relu")(e)
e = Model(inputs=inputs_7, outputs=e)


f = Dense(8, activation="relu")(inputs_8)
f = Dense(4, activation="relu")(f)
f = Model(inputs=inputs_8, outputs=f)

g = Dense(8, activation="relu")(inputs_9)
g = Dense(4, activation="relu")(g)
g = Model(inputs=inputs_9, outputs=g)

# combine the output of the branches
combined = Concatenate()([x.output, y.output, a.output, b.output, c.output, d.output, e.output, f.output, g.output])

'''
# apply a FC layer and then a regression prediction on the
# combined outputs
#z = Dense(2, activation="relu")()
#z = Dense(1, activation="linear")(z)

y_output = Dense(units='1')(x)

# our model will accept the inputs of the two branches and
# then output a single value
#model = Model(inputs=[x.input, y.input, a.input, b.input, c.input, d.input, e.input, f.input, g.input], outputs=z)
model = Model(inputs=inputs,outputs=y_output)
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
print(np.array(input_x))
for layer in model.layers:
    print(layer.get_output_at(0).get_shape().as_list())
model.fit(train_X, train_Y, epochs=150, batch_size=10, verbose=0)
```
